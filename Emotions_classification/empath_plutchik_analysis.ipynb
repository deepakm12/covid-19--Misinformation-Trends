```python
a = plutchik[plutchik["joy", "sadness" , "trust", "surprise", "fear", "disgust", "anticipation", "anger"].astype(bool)]
```

```python
#total tweets classified by plutchik
coverage = plutchik[plutchik["joy"].astype(bool) | plutchik["sadness"].astype(bool) | plutchik["trust"].astype(bool) | plutchik["fear"].astype(bool) | plutchik["anger"].astype(bool) | plutchik["disgust"].astype(bool) | plutchik["anticipation"].astype(bool) | plutchik["surprise"].astype(bool)]
```

```python
#total tweets classified by both plutchik and empath
merged_df = plutchik.merge(empath, on='id_str', how='inner', suffixes=('_plutchik', '_empath'))
```

```python
#percentage total no. tweets classified by  plutchik 
t = (merged_df['joy_empath'] == merged_df['joy_plutchik']).sum()/len(merged_df)
t
```

```python
#finding common classified by both plutchik and empath in each emotion categories
emotions = ["joy", "sadness" , "trust", "surprise", "fear", "disgust", "anticipation", "anger"]
for emotion in emotions:
    empath_col = f'{emotion}_empath'
    plutchik_col = f'{emotion}_plutchik'
    print(emotion, (merged_df[empath_col] == merged_df[plutchik_col]).sum()/len(merged_df))
```

```python
 #tweets classified by plutchik but no in empath
not_in_empath = coverage.merge(empath,indicator = True, how='left', on=['id_str']).loc[lambda x : x['_merge']!='both']
not_in_empath.head(100)
```

```python
import gzip
import shutil
import os
import wget
import csv
import linecache
from shutil import copyfile
import numpy as np
import pandas as pd
import time
```

```python
#genrating date
li = []
for i in range(21, 32):
  if(i<10):
    li.append("0"+str(i))
  else:
    li.append(str(i))
```

```python
#data downloading
for date in li:
  date = str(date)
  dataset_URL = "https://github.com/thepanacealab/covid19_twitter/blob/master/dailies/2020-05-"+date+"/2020-05-"+date+"_clean-dataset.tsv.gz?raw=true"
  # os.mkdir('/content/drive/MyDrive/Covid19/May2020/'+date+'may')
  wget.download(dataset_URL, out='/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset.tsv.gz')

  with gzip.open('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset.tsv.gz', 'rb') as f_in:
      with open('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset_1.tsv', 'wb') as f_out:
          shutil.copyfileobj(f_in, f_out)

  os.unlink('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset.tsv.gz')


  #english language filtering and location filtering
  filtered_language = 'en'
  filtered_tw = list()
  current_line = 1
  with open('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset_1.tsv') as tsvfile:
    tsvreader = csv.reader(tsvfile, delimiter="\t")

    if current_line == 1:
      filtered_tw.append(linecache.getline('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset_1.tsv', current_line))

      for line in tsvreader:
        if current_line>10000:
          filtered_tw.append(linecache.getline('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset_1.tsv', current_line))
        current_line += 1
        if current_line==60000:  break
    with open('/content/drive/MyDrive/Covid19/May2020/'+date+'may'+'/dataset_1.tsv', 'w') as f_output:
        for item in filtered_tw:
            f_output.write(item)
  
```

```python
#getting feb 2021 dates
li = []
for i in range(1, 29):
  if(i<10):
    li.append("0"+str(i))
  else:
    li.append(str(i))
```

```python
#data downloading
for date in li:
  date = str(date)
  dataset_URL = "https://github.com/thepanacealab/covid19_twitter/blob/master/dailies/2021-02-"+date+"/2021-02-"+date+"_clean-dataset.tsv.gz?raw=true"
  os.mkdir('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb')
  wget.download(dataset_URL, out='/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv.gz')

  with gzip.open('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv.gz', 'rb') as f_in:
      with open('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv', 'wb') as f_out:
          shutil.copyfileobj(f_in, f_out)

  os.unlink('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv.gz')


  #english language filtering and location filtering
  filtered_language = 'en'
  filtered_tw = list()
  current_line = 1
  with open('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv') as tsvfile:
    tsvreader = csv.reader(tsvfile, delimiter="\t")

    if current_line == 1:
      filtered_tw.append(linecache.getline('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv', current_line))

      for line in tsvreader:
        if line[3] == filtered_language and line[4]!="NULL":
          filtered_tw.append(linecache.getline('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv', current_line))
        current_line += 1
        # if current_line==200:  break
    with open('/content/drive/MyDrive/Covid19/Feb2021/'+date+'feb'+'/dataset.tsv', 'w') as f_output:
        for item in filtered_tw:
            f_output.write(item)
  
```

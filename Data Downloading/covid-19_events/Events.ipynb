```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
response = requests.get('https://www.thinkglobalhealth.org/article/updated-timeline-coronavirus')
```

```python
response = BeautifulSoup(response.text, 'html.parser')
```

```python
response.body.string
```

```python
mylist = []
for article in response.find_all('div', {'class': 'o-article-body__content'}):
  for mlist in article.find_all('ul'):
    temp = []
    for mmlist in mlist.find_all('li'):
      temp.append(mmlist.text.strip())
    mylist.append(temp)
```

```python
mylist = list(filter(None, mylist))
```

```python
data = pd.DataFrame(columns = ['date','event'])
```

```python
date = []
for article in response.find_all('div', {'class': 'o-article-body__content'}):
  for mlist in article.find_all('p'):
    if mlist.strong:
      date.append(mlist.strong.string.strip())
      
```

```python
date.insert(470,'February 24')
```

```python
date.insert(496,'January 29')
```

```python
for i in range(len(date)):
  for mli in mylist[i+1]:
    data = data.append({'date':date[i],'event':mli}, ignore_index=True)
    print(date[i])
```
